```python
import numpy as np
```

```python
GHz = 1e9
c =3e8
```

```python
N = 8001

f0 = 50*GHz
BW = 40*GHz

factor0 = 4*np.pi/c
space_resol = 1e-2
din, df = space_resol, (N-1)*c/BW
gamma_0, gamma_1 = -2, 2
gamma_resol = 0.1

d_vals = np.linspace(din, df, np.ceil(int((df-din)/space_resol)))
gamma_vals = np.linspace(gamma_0, gamma_1, int((gamma_1 - gamma_0)/gamma_resol))

ff = f0 + BW

n = np.linspace(0, N-1, N)
```

```python
def h_time_domain(d_, gamma_, n):
    
    h = 0
    alfa0 = d_*factor0*(BW/(N-1))
    beta0 = d_*factor0*f0
    
    for k in range(N):
        h += (1/(alfa0*k+beta0)**gamma_)*np.exp(-1j*(alfa0*k+beta0)/2)*np.exp(1j*2*np.pi*k*n/N)
        #h += (1/(alfa*k+beta)**gamma)*np.exp(1j*2*np.pi*k*n/N)
    return h/N
```

```python
H = []
i=0
for d in np.linspace(1e-2, df, N):
    h_d = h_time_domain(d, 1, n)
    H.append(h_d)
    i+=1
    print(f'Processing function {i}/{N}')
```

```python
def scalar_product(h1, h2):

    return np.sum(h1*np.conj(h2))

def module(h):

    return np.sqrt(scalar_product(h, h))
```

```python
H = np.array(H)
```

```python
for i in range(len(H)):  
    print(f'Processing function {i} out of {len(H)}')
    H[i] = H[i]/module(H[i])
```

```python
np.savez(f'H_distance_8001_f0={f0/GHz}_BW={BW/GHz}.npz', x=H)
```
